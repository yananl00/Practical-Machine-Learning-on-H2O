```python
import h2o
```

```python
h2o.init()
```

```python
url = "http://h2o-public-test-data.s3.amazonaws.com/smalldata/iris/iris_wheader.csv"
iris = h2o.import_file(url)
```

```python
train, test = iris.split_frame([0.8])
```

Random Forest

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator

mRF = H2ORandomForestEstimator()
mRF.train(["sepal_len", "sepal_wid", "petal_len", "petal_wid"],"class", train)
```

```python
mRF
```

```python
p = mRF.predict(test)
```

```python
mRF.model_performance(test)
```

## Train, valid, test

```python
people = h2o.get_frame("people")

train, valid, test = people.split_frame(
    ratios = [0.8, 0.1],
    destination_frames = ["people_train", "people_valid", "people_test"], # send to flow
    seed = 123)
```

```python
train = h2o.get_frame("people_train")
valid = h2o.get_frame("people_valid")
test = h2o.get_frame("people_test")
```

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator

y = "income"
ignoreFields = [y, "id"]
x = [i for i in train.names if i not in ignoreFields]

m1 = H2OGradientBoostingEstimator(model_id = "defaults")
m1.train(x, y, train, validation_frame = valid)
```

```python
m1.mae(train = True)
```

```python
m1.mae(valid = True)
```

```python
perf = m1.model_performance(test)
perf.mae()
```

```python
m2 = H2OGradientBoostingEstimator(model_id = "overfit", ntrees = 1000, max_depth = 10)
m2.train(x, y, train, validation_frame = valid)
```

## Early stopping to avoid overfitting, scores are evaluated on validation/cross validation dataset, rather than the training dataset

```python
m2 = H2OGradientBoostingEstimator(model_id = "early_stop", 
                                  stopping_rounds = 4, 
                                  stopping_metric = "AUTO", 
                                  stopping_tolerance = 0.001,
                                  ntrees = 1000, 
                                  max_depth = 10)

m2 = H2OGradientBoostingEstimator(model_id = "mae_worsens", 
                                  stopping_rounds = 4, 
                                  stopping_metric = "MAE", 
                                  stopping_tolerance = 0.001,
                                  ntrees = 1000, 
                                  max_depth = 10)

m2.train(x, y, train, validation_frame = valid)
```

## Cross validation

```python
people = h2o.get_frame("people")

train, test = people.split_frame(
    ratios = [0.897],
    destination_frames = ["people_train", "people_test"], # send to flow
    seed = 123)
```

```python
m3 = H2OGradientBoostingEstimator(model_id = "def9folds", nfolds = 9)
m3.train(x, y, train)
```

```python
m3.mae(train = True)
```

```python
m3.mae(xval = True)
```

```python
perf = m3.model_performance(test)
perf.mae()
```
